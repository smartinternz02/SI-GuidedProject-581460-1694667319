OUTLIER DETECTION

```python
sns.boxplot(x=df["Age"])
plt.show()
```

```python
# Define a function to remove outliers using IQR method
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Specify columns for outlier detection (e.g., 'Fare' and 'Age')
columns_to_check = ['Fare', 'Age']

# Apply outlier removal function to specified columns
for column in columns_to_check:
    df = remove_outliers_iqr(df, column)

df.reset_index(drop=True, inplace=True)
```

```python
# Define a function to remove outliers using Z-score method 
from scipy.stats import zscore

# Define a function to remove outliers using Z-score method
def remove_outliers_zscore(data, column, threshold=3):
    z_scores = zscore(data[column])
    outliers = (abs(z_scores) > threshold)
    return data[~outliers]

# Specify columns for outlier detection (e.g., 'Fare' and 'Age')
columns_to_check = ['Fare', 'Age']

# Apply outlier removal function to specified columns
for column in columns_to_check:
    df = remove_outliers_zscore(df, column)

# Reset index after removal
df.reset_index(drop=True, inplace=True)

# Now titanic_df contains the dataset with outliers removed using Z-score method
```
